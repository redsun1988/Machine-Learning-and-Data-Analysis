```python
import scipy
import math
import numpy as np
from scipy import optimize
from scipy.stats import powerlaw # Импортируем распределение, имеющее название powerlaw - частный случай бета-распределения с b=1
from scipy.stats import norm
import matplotlib.pyplot as plt

a = 3.99 # Задаем параметр a для распределения
##fig, ax = plt.subplots(1, 1)
xx = np.linspace(0,2,1000)
samples_pl = powerlaw.rvs(a, size = 1000) # Отобразим исходное распределение
plt.plot(xx, powerlaw.pdf(xx,a),'r-', lw=2, label='PDF with a=3.99', alpha = 0.6)
plt.hist(samples_pl, bins= 50, normed=True, histtype='stepfilled', alpha=0.2, label='Hist w/step=50')
plt.legend(loc='best', frameon=False)
plt.title("Powerlaw")
plt.ylabel('$PDF$')
plt.xlabel('$x$')
plt.show() # Рисуем необходимые графики и гистограммы

mu_t = powerlaw.mean(a) # Теоретическое среднее значение исходного распределения
sigma_t = np.sqrt(powerlaw.var(a)) # Теоретическое отклонение исходного распределения

rndm5 = powerlaw.rvs(a, size = 1000) # Оценим распределение выборочного среднего для n=5
means5 = []
for i in range(1000):
    slice5 = rndm5[np.random.randint(0,len(rndm5),5)] #выбираем 5 значений
    means5.append(np.mean(slice5)) #Наполняем массив выборочных средних значением для текущих 5
mu5 = np.mean(means5) # Измеренное среднее значение
sigma5 = np.std(means5,ddof=1) # Измеренное отклонение на основаннии выборки
SE5 = sigma_t/np.sqrt(5) # Теоретическое отклонение задается исходя из ЦПТ
print('При n=5: ')
print('Теоретическое среднее:', mu_t)
print('   Измеренное среднее:', mu5)
print('Теоретическое отклонение:', SE5)
print('   Измеренное отклонение:', sigma5)
norm_n5 = norm(mu_t,SE5) # Нормальное распределение с заданными параметрами
plt.plot(xx, norm_n5.pdf(xx), 'r-', alpha=0.5, label='Теоретическое распределение')
plt.hist(means5, normed=True, alpha=0.2, label='Измеренные средние')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., frameon=False)
plt.title("n=5")
plt.ylabel('$PDF$')
plt.xlabel('$x$')
plt.show() # Рисуем необходимые графики и гистограммы

rndm10 = powerlaw.rvs(a, size = 1000) # Оценим распределение выборочного среднего для n=10
means10 = []
for i in range(1000):
    slice10 = rndm10[np.random.randint(0,len(rndm10),10)] #выбираем 10 значений
    means10.append(np.mean(slice10)) #Наполняем массив выборочных средних значением для текущих 10
mu10 = np.mean(means10) # Измеренное среднее значение
sigma10 = np.std(means10,ddof=1) # Измеренное отклонение на основаннии выборки
SE10 = sigma_t/np.sqrt(10) # Теоретическое отклонение задается исходя из ЦПТ
print('При n=10: ')
print('Теоретическое среднее:', mu_t)
print('   Измеренное среднее:', mu10)
print('Теоретическое отклонение:', SE10)
print('   Измеренное отклонение:', sigma10)
norm_n10 = norm(mu_t,SE10) # Нормальное распределение с заданными параметрами
plt.plot(xx, norm_n10.pdf(xx), 'r-', alpha=0.5, label='Теоретическое распределение')
plt.hist(means10, normed=True, alpha=0.2, label='Измеренные средние')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., frameon=False)
plt.title("n=10")
plt.ylabel('$PDF$')
plt.xlabel('$x$')
plt.show() # Рисуем необходимые графики и гистограммы

rndm50 = powerlaw.rvs(a, size = 1000) # Оценим распределение выборочного среднего для n=50
means50 = []
for i in range(1000):
    slice50 = rndm5[np.random.randint(0,len(rndm50),50)] #выбираем 50 значений
    means50.append(np.mean(slice50)) #Наполняем массив выборочных средних значением для текущих 50
mu50 = np.mean(means50) # Измеренное среднее значение
sigma50 = np.std(means50,ddof=1) # Измеренное отклонение на основаннии выборки
SE50 = sigma_t/np.sqrt(50) # Теоретическое отклонение задается исходя из ЦПТ
print('При n=50: ')
print('Теоретическое среднее:', mu_t)
print('   Измеренное среднее:', mu50)
print('Теоретическое отклонение:', SE50)
print('   Измеренное отклонение:', sigma50)
norm_n50 = norm(mu_t,SE50) # Нормальное распределение с заданными параметрами
plt.plot(xx, norm_n50.pdf(xx), 'r-', alpha=0.5, label='Теоретическое распределение')
plt.hist(means50, normed=True, alpha=0.2, label='Измеренные средние')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., frameon=False)
plt.title("n=50")
plt.ylabel('$PDF$')
plt.xlabel('$x$')
plt.show() # Рисуем необходимые графики и гистограммы
# Хорошо видно как по мере увеличения размера выборки среднее все больше и больше начинает напоминать нормальное распределение
# Одновременно с этим и увеличивается точность аппроксимации к теоретическим значениям.
```
