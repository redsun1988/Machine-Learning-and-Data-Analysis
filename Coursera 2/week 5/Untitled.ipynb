```python
from sklearn import datasets, cross_validation, metrics 
import math as math 

import numpy as np
import pandas as pd
from functools import reduce
```

```python
digits = datasets.load_digits()
X = digits.data
Y = digits.target
```

```python
def write_answer(fileName, value):
    with open(fileName+".txt", "w") as fout:
        fout.write(str(value))
```

```python
def calcDistance(row1, row2):
    result = 0
    for i, value in enumerate(row1):
        result += (value - row2[i])**2
    return math.sqrt(result)
```

```python
#X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=0.25)

X_train  = X[:int((len(X)*0.75))]
X_test = X[-(int((len(X)*0.25)) + 1):]
Y_train = Y[:int((len(Y)*0.75))]
Y_test = Y[-(int((len(Y)*0.25)) + 1):]
```

```python
def predefine(rowForCheck):
    distances = []
    for row in X_train:
        distances.append(calcDistance(row, rowForCheck))
    return Y_train[np.argmin(distances)]
```

```python
falseCount = 0 
for i, row in enumerate(X_test):
    if predefine(row) != Y_test[i]:
        falseCount += 1
print falseCount
```

```python
float(falseCount) / X_test.shape[0]
```

```python
write_answer("part3_task1", float(falseCount) / X_test.shape[0])
```

```python
from sklearn import ensemble
rf_classifier = ensemble.RandomForestClassifier(n_estimators = 1000)
rf_classifier.fit(X_train, Y_train)
```

```python
rf_falseCount = 0 
for i, row in enumerate(X_test):
    if rf_classifier.predict(row)[0] != Y_test[i]:
        rf_falseCount += 1
print rf_falseCount

result  = rf_classifier.predict(X_test[0])
```
